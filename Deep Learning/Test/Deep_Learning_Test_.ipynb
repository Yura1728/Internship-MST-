```python
''' 1. A Multi-Layer Perceptron (MLP) is a type of artificial neural network that consists of multiple layers of perceptrons.
 It has an input layer, one or more hidden layers, and an output layer
 MLPs are used for more complex problems like image recognition
 MLPs are more complex and can learn non-linear functions
 But in Single layer Perceptron we have input layer and one hidden layer and output layer
 In single-layer perceptrons can only learn linear functions'''
```

```python
''' 2. The Architecture of a Multi-Layer Perceptron (MLP) is
The input layer is the first layer of the MLP and receives the input data.
The hidden layers are the core of the MLP
Each node in the hidden layer applies an activation function to the weighted sum of its inputs from the previous layer.
And These are Dense Layers or Fully Connected Layers
And weights are adjusted during training to minimize the error between the predicted and actual outputs.
The output layer is the final layer of the MLP and produces the predicted output'''
```

```python
''' 3. Weights in a Multi-Layer Perceptron (MLP) are initialized using various techniques
The initialization of weights is crucial because it directly affects the learning process and the final performance of the model
To reduce the loss and to gain more Accuracy we update the weights, using the below techniques
Like He Initialization, Xavier Initialization etc.
And weights are updated to reduce the Vanishing Gradient Problem and Exploding Gradient Problem'''
```

```python
''' 4. The purpose of activation functions in a Multi-Layer Perceptron (MLP) is to learn and represent complex relationships in the data.
Activation functions are used in the hidden and output layers of the MLP to transform the weighted sum of the inputs
into an output that can be used by the next layer.
Some Activation Functions are ReLu, SoftMax, Sigmoid, TanH, Leaky ReLu, Exponential ReLu and Swish function'''
```

```python
''' 5. Backpropagation is a method used to train artificial neural networks, including Multi-Layer Perceptrons (MLPs), by minimizing the errors
between the predicted output and the actual output.
It is based on the chain rule and is used to compute the gradients of the loss function with respect to the model's parameters.
The backpropagation is used to train MLPs by iteratively adjusting the weights and biases to minimize the loss function.
The gradients are then used to update the weights and biases using gradient descent or other optimization algorithms.'''
```

```python
'''6. Choosing the number of hidden layers and neurons in each layer of a Multi-Layer Perceptron (MLP) is crucial for the performance of the model.
We can Start with One Hidden Layer, and then add more hidden layers if necessary.
This is because a single hidden layer can learn complex relationships in the data.
If the data requires more complex transformations, you can add more hidden layers.
And we should be do it with caution as too many hidden layers can lead to overfitting of the model
So actually we can choose the optimal number of hidden layers and neurons for your Multi-Layer Perceptron (MLP) model. Based on your requirements and the complexity of the data'''
```

# Load MNIST Dataset and build a Deep Learning Model.

### Things to be done:
    1- Load the MNIST Data
    2- Divide the data into train and test sets
    3- Plot the first 10 images using imshow()
    4- Convert images to categorical format
    5- Create a sequential model
    6- Plot the model summary
    7- Plot the architecture of the model with its layer shapes
    8- Compile the model
    9- Fit the model with 20 epochs (as per your desktop configuration)
    10- Plot Training and Testing Loss
    11- Save the model in .h5 extension
    12- Load the model
    13- Test the model

```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical
from keras import utils
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
fig, axes = plt.subplots(ncols=10, sharex=False, sharey=True, figsize=(20, 4))
for i in range(10):
    axes[i].set_title(y_train[i])
    axes[i].imshow(x_train[i], cmap='gray')
    axes[i].get_xaxis().set_visible(False)
    axes[i].get_yaxis().set_visible(False)
plt.show()
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
print(y_train)
print(y_test)
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(10, activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
utils.plot_model(model, show_shapes=True)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
```

```python
model.save('mnist_model.h5')
```
