```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=['Car','Bike','Auto']
y=[12,22,35]
plt.bar(x,y)
plt.show()
```

```python
x=np.array(['Pen','Pencil','Colour Pencil'])
y=np.array([5,6,2])
plt.barh(x,y)
plt.show()
```

```python
y=np.array([10,15,25,36])
mylabels=['Car','Bike','Auto','Bus']
myexplode=[0.3,0.2,0.4,0.1]
mycolor=['Orange','Blue','Green','Black']
plt.pie(y, labels=mylabels, startangle=90, explode=myexplode, colors=mycolor)
plt.legend(title='Vehicles')
plt.show()
```

```python
x=np.array([11,12,13,15,18,19,25,31,46,52])
y=np.array([32,54,76,28,65,98,75,48,23,29])
plt.scatter(x,y,alpha=0.4)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y1 = [38, 48, 48, 50]
y2 = [29, 25, 35, 47]
plt.bar(x, y1)
plt.bar(x, y2)
plt.xlabel("Sections")
plt.ylabel("Marks")
plt.title('Stack Bar chat')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [3, 3, 3, 3, 3]
y2 = [4, 5, 6, 8, 10]
y3 = [5, 4, 3, 2, 1]
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.show()
```

```python
df=pd.read_csv('/content/electronic-card-transactions-january-2024-csv-tables.csv')
plt.bar(df['UNITS'],df['STATUS'])
plt.title('Bar chat using CSV')
plt.xlabel('Units')
plt.ylabel('Status')
plt.show()
```

```python
plt.barh(df['UNITS'],df['STATUS'])
plt.title('Horizontal Bar chat using CSV')
plt.xlabel('Status')
plt.ylabel('Units')
plt.show()
```
