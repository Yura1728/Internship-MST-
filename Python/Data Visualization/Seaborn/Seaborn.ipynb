```python
import seaborn as sns
sns.set(style="dark")
fmri=sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",hue="region",style="event",data=fmri)
```

```python
df=sns.load_dataset('tips')
sns.barplot(x='sex',y='total_bill',data=df,palette ='plasma')
```

```python
df=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker')
```

```python
df=sns.load_dataset('tips')
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True)
```

```python
sns.scatterplot(x='day',y='tip',data=df,marker='+')
```

```python
sns.scatterplot(x='day',y='tip',data=df,hue='time')
```

```python
sns.scatterplot(x='day',y='tip', data=df,hue='time',palette='pastel')
```

```python
import numpy as np
sns.set(style='white')
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.histplot(d, kde=True, color='m')
```

```python
sns.set(style='white')
rs=np.random.RandomState(10)
d=rs.normal(size=100)
colors=['r', 'g', 'b']
sns.distplot(d, kde=True, hist=True, bins=10)
```
